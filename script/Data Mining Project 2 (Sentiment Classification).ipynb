```python
def retrieveListsOfTweets(filePath):
    data = pd.ExcelFile(filePath)
    #print(data.sheet_names)
    sheet1_data = data.parse('Obama')
    sheet2_data = data.parse('Romney')
    sheet1_data_tweet_column = sheet1_data.iloc[1:,3:5]
    sheet2_data_tweet_column = sheet2_data.iloc[1:,3:5]
    list_Obama_tweets = []
    list_Romney_tweets = []
    for row in sheet1_data_tweet_column.itertuples():
        list_Obama_tweets.append((row[1],row[2]))
    for row in sheet2_data_tweet_column.itertuples():
        list_Romney_tweets.append((row[1],row[2]))
    return list_Obama_tweets,list_Romney_tweets

```

```python
def dataPreProcessing(rawData):
    tweets = []
    #remove hyperlinks, html tags
    for row in rawData:
        tweet = re.sub(r'https?://.*[\r\n]*','', row[0])
        tweet = re.sub(r'<..?>','',tweet)
        print(str(tweet))
        tweets.append(tweet)
    
    return tweets
```

```python
dataPreProcessing(Obama_tweets)
```
